```python
char_cols =df.dtypes.pipe( lambda x:x[x=="object"]).index
label_mapping={}
for c in char_cols:
    df[c],label_mapping[c]=pd.factorize(df[c])
```

```python
df= df.drop(columns=["capital-gain","capital-loss"])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df = scaler.fit_transform(df.values)
```

```python
#split the data
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(df,target)
```

```python
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
```

```python
clf= LogisticRegression().fit(x_train,y_train)
```

```python
pred= clf.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier().fit(x_train, y_train)
pred_dt= model.predict(x_test)
```

```python
print(classification_report(y_test,pred_dt))
```

```python
from sklearn.neighbors import KNeighborsClassifier
modell= KNeighborsClassifier().fit(x_train, y_train)
pred_knn= modell.predict(x_test)
```

```python
print(classification_report(y_test,pred_knn))
```

```python
from sklearn.ensemble import RandomForestClassifier
modrc=RandomForestClassifier().fit(x_train, y_train)
predkrc= modrc.predict(x_test)
```

```python
print(classification_report(y_test,predkrc))
```

```python
from sklearn.ensemble import ExtraTreesClassifier
modet=ExtraTreesClassifier().fit(x_train, y_train)
predet= modet.predict(x_test)
```

```python
print(classification_report(y_test,predet))
```
